# Histogram
In matplotlib we can create a histogram using the hist method . if we pass it categorical data like the points column from the wine-review dataset it will automatically calculate how often each class occurs

```python
#create a figure and axis
fig, ax = plt.subplots()
# plot histogram
ax.hist(iris['SepalLengthCm'])
#set title and label
ax.set_title('iris')
ax.set_xlabel('SepalLengthCm')
ax.set_ylabel('Frequency')
```

# Bar Chart

```python
wine_reviews = pd.read_csv("C:/Users/Saroj/Downloads/winemag-data-130k-v2.csv")
wine_reviews.head()
```

```python
#create a figure and axis
fig, ax = plt.subplots()
#count the occurence of each class
data = wine_reviews['points'].value_counts()
# get x and y data
points = data.index
frequency = data.values
# create bar chart
ax.bar(points, frequency)
ax.set_title('wine_reviews') 
ax.set_xlabel('Points')
ax.set_ylabel('Frequency')
```

```python
import numpy as np
import matplotlib.pyplot as plt
objects = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'lisp')
y_pos = np.arange(len(objects))
performance = [10,8,6,4,2,1]
#bar chart
#x axis positions as first parameter list, it can be floating points numbers also
# y alues as 2nd parameter list
# alphais a transparancy,
#coor can be single value or a list of color codes, one for each bar.
plt.bar(y_pos, performance, width=0.5, align='center', alpha=0.7, color=['r', 'r', 'g', 'g', 'b','b'])
# to define lable for x axis values.
plt.xticks(y_pos, objects)
plt.ylabel('Usage')
plt.xlabel('Programming Languages')
plt.title('Programming languages usage')
```

```python
# to print the values above the bar
import matplotlib.pyplot as plt

# Declaring the figure or the plot (y,x) or (width, height)
plt.figure(figsize=(12, 7))

# Categorical data: Country names
countries = ['USA', 'Brazil', 'Russia', 'Spain', 'UK', 'India']

# integer value interms of death counts
totalDeaths = [112596, 37312, 5971, 27136, 40597, 7449]

# Passing the parameters to the bar function, this is the main function which creates the bar plot
plt.bar(countries, totalDeaths, width=0.9, align='center', color='cyan', edgecolor = 'red')

# This is the location for the annotated text
i =1.0
j =2000

# Annotating the bar plot with the values (total death counts)
for i in range(len(countries)):
    plt.annotate(totalDeaths[i], (-0.1+ i, totalDeaths[i] + j))

# Creating legend of the bars in the plot
plt.legend(labels = ['Total Deaths'])

# Giving the title for the plot
plt.title('Bar plot representing the total deaths by top 6 countries due to coronavirus')

# Naming the x and y axis
plt.xlabel('Countries')
plt.ylabel('Deaths')

# Saving the plot as a 'png'
plt.savefig("BarPlot.png")

# Display the bar plot
plt.show()
```
