# Heatmap

```python
data=np.random.randint(1,100,size=(10,10))
print("The data to be plotted: \n")
print(data)
```

```python
hm=sns.heatmap(data=data)
plt.show()
```

```python
hm=sns.heatmap(data=data)
plt.show()
```

```python
hm = sns.heatmap(data=data,
                 vmin='30',
                 vmax='70')
plt.show()
```

```python
#setting the parameter values
cmap = "tab20"
center = 0

annot = True
hm = sns.heatmap(data=data, cmap=cmap, annot=annot)
plt.show()
```
